```python
import threading
```

```python
import requests
import time
import os
import shutil
```

```python
file_urls = [
    "https://www.w3.org/WAI/ER/tests/xhtml/testfiles/resources/pdf/dummy.pdf",
    "https://www.learningcontainer.com/wp-content/uploads/2020/05/sample-zip-file.zip",
    "https://www.learningcontainer.com/wp-content/uploads/2020/04/sample-text-file.txt"
]
```

```python
folder = "downloads"

if not os.path.exists(folder):
    os.makedirs(folder)
```

```python
def download_file(url):
    try:
        file_name = url.split("/")[-1]
        file_path = os.path.join(folder, file_name)

        print(f"{threading.current_thread().name} is downloading {file_name}...")

        response = requests.get(url)

        with open(file_path, "wb") as file:
            file.write(response.content)

        print(f" {file_name} downloaded successfully!")

    except Exception as e:
        print(f" Error downloading {url}")
        print(e)
```

```python
def single_thread_download():

    clear_download_folder()

    start_time = time.time()

    for url in file_urls:
        download_file(url)

    end_time = time.time()

    total = end_time - start_time

    print(f"\nSingle Thread Time : {total:.2f} seconds")

    return total
```

```python
def multi_thread_download():
    threads = []

    start_time = time.time()

    for url in file_urls:
        thread = threading.Thread(target=download_file, args=(url,))
        threads.append(thread)
        thread.start()

    for thread in threads:
        thread.join()

    end_time = time.time()

    print(f"\nMulti Thread Time: {end_time - start_time:.2f} seconds")
```

```python
def clear_download_folder():

    if os.path.exists(folder):
        shutil.rmtree(folder)

    os.makedirs(folder)

    print("Downloads folder cleared.\n")
```

```python
def multi_thread_download():

    clear_download_folder()

    threads = []

    start_time = time.time()

    for url in file_urls:

        thread = threading.Thread(target=download_file, args=(url,))

        threads.append(thread)

        thread.start()

    for thread in threads:
        thread.join()

    end_time = time.time()

    total = end_time - start_time

    print(f"\nMulti Thread Time : {total:.2f} seconds")

    return total
```

```python
start_time = time.time()
```

```python
threads=[]
for url in file_urls:
  thread = threading.Thread(target=download_file, args=(url,))
  threads.append(thread)
  thread.start()
```

```python
for thread in threads:
  thread.join()

end_time = time.time()

print("\n All downloads completed!")
print(f"Total time taken: {end_time - start_time:.2f} seconds")
```

```python
print("===== SINGLE THREAD =====")

single_time = single_thread_download()

print("\n============================\n")

print("===== MULTI THREAD =====")

multi_time = multi_thread_download()

print("\n============================")

speedup = single_time / multi_time

print(f"\n🚀 Speed Improvement : {speedup:.2f}x Faster")
```
